```python
def get_test(code, code_value, subject):
    count = 0
    total = 0
    for student in student_info:
        if student[prop_keys[code]] == code_value: # student[prop_keys[code]] -> student_info에서 student 행에 code_value가 입력값과 같다면
            total = total + student[prop_keys[subject]] # 총합에 넣는다
            count = count +1 # 카운트 1 올린다
            pass
        pass
    return total/count # 조건을 만족한 값의 총합 / 조건을 만족한 행의 수
```

```python
# 강사님 방법
# sort = 0 (오름차순) sort = 1 (내림차순) start = 1 count = 1
def get_select_with_sort(target, sort=0, start=1, count=1):
    print('sort',sort)
    result = None
    for idx1 in range(len(student_info)-1):
        for idx2 in (idx1+1,len(student_info)):
            if sort == 0:
                if student_info[idx1][prop_keys[target]] > student_info[idx2][prop_keys[target]]:
                    temp = student_info[idx1]
                    student_info[idx1] = student_info[idx2]
                    student_info[idx2] = temp
                    pass
            else:
                    if student_info[idx1][prop_keys[target]] < student_info[idx2][prop_keys[target]]:
                        temp = student_info[idx1]
                        student_info[idx1] = student_info[idx2]
                        student_info[idx2] = temp
                    pass
                
    count = start- 1 + count
    print('end',count)
    result = student_info[start-1:count]                           
    
    return result
```
